```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
df = pd.read_csv(r'C:\Users\ntpc\Desktop\Movie_classification.csv')
df.head()
```

## Missing Value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df,columns = ["3D_available","Genre"],drop_first = True)
```

## X-y split

```python
X = df.loc[:,df.columns!="Start_Tech_Oscar"]
```

```python
y = df["Start_Tech_Oscar"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
X_test_std
```

```python
from sklearn import svm
```

```python
clf_svm_l = svm.SVC(kernel='linear', C=100)
clf_svm_l.fit(X_train_std, y_train)
```

```python
y_train_pred = clf_svm_l.predict(X_train_std)
y_test_pred = clf_svm_l.predict(X_test_std)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```
